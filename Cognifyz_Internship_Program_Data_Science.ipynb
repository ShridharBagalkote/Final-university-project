```python
# Level-1
 # Task-:  Data Exploration and Preprocessing
 # Task 1
 # Explore the dataset and identify the number of rows and columns.
 # Check for missing values in each column and
 # handle them accordingly.
 # Perform data type conversion if necessary.
 # Analyze the distribution of the target variable
 # ("Aggregate rating") and identify any class
 # imbalances.

import pandas as pd

# 1. Load the Dataset
df = pd.read_csv(r"/Dataset .csv")
df
```

```python
# 2. Explore Rows and Columns
num_rows, num_cols = df.shape
print("Number of rows:", num_rows)
print("Number of columns:", num_cols)
```

```python
# 3. Check for Missing Values
missing_values = df.isnull().sum()
print("\nMissing values per column:\n", missing_values)
```

```python
# 5. Data Type Conversion
#    Inspect data types:
print("\nData Types:\n", df.dtypes)
#    Convert 'Average Cost for two' to float (if it was loaded as a string):
df['Average Cost for two'] = df['Average Cost for two'].astype(float)
```

```python
# 6. Target Variable Distribution
rating_counts = df['Aggregate rating'].value_counts().sort_index()
print("\n'Aggregate rating' distribution:\n", rating_counts)
```

```python
# 7. Identify Class Imbalances
#    Calculate the percentage of each rating class:
class_percentages = rating_counts / len(df) * 100
print("\nClass Percentages:\n", class_percentages)
```

```python
# 8   Visualize the distribution (helpful for spotting imbalances):
import matplotlib.pyplot as plt
plt.bar(class_percentages.index, class_percentages.values)
plt.xlabel('Aggregate Rating')
plt.ylabel('Percentage')
plt.title('Distribution of Aggregate Rating')
plt.show()
```

```python
#Task:2  Descriptive Analysis
 # Calculate basic statistical measures (mean,
 # median, standard deviation, etc.) for numerical columns.
 # Explore the distribution of categorical
 # variables like "Country Code," "City," and "Cuisines."
 # Identify the top cuisines and cities with the
 # highest number of restaurants

import pandas as pd

# Load the dataset
df = pd.read_csv(r"/Dataset .csv")

# Descriptive analysis for numerical columns
numerical_columns = ['Longitude', 'Latitude', 'Average Cost for two', 'Price range', 'Aggregate rating', 'Votes']
df[numerical_columns].describe()

# Distribution of categorical variables
categorical_columns = ['Country Code', 'City', 'Cuisines']
for col in categorical_columns:
  print(f"Distribution of {col}:\n{df[col].value_counts()}\n")

# Top cuisines and cities with the highest number of restaurants
print(f"Top cuisines:\n{df['Cuisines'].value_counts().nlargest(10)}\n")
print(f"Cities with the highest number of restaurants:\n{df['City'].value_counts().nlargest(10)}")




```

```python
 #Task:  Geospatial Analysis
 # Task 3
 # Visualize the locations of restaurants on a
 # map using latitude and longitude information.
 # Analyze the distribution of restaurants across different cities or countries.
 # Determine if there is any correlation between the restaurant's location and its rating

import pandas as pd
import plotly.express as px

# Load the dataset
df = pd.read_csv(r"/Dataset .csv")

# Create a scatter mapbox using plotly.express
fig = px.scatter_mapbox(
    df,
    lat="Latitude",
    lon="Longitude",
    hover_name="Restaurant Name",
    hover_data=["Cuisines", "Average Cost for two", "Aggregate rating", "Rating text"],
    color="Rating color",
    color_discrete_map={
        "Dark Green": "green",
        "Green": "lightgreen",
        "Orange": "orange",
        "Red": "red",
        "White": "gray"
    },
    zoom=2,
    height=600,
    width=800
)

# Customize the map appearance
fig.update_layout(mapbox_style="open-street-map")
fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})

# Show the map
fig.show()

# Correlation analysis (example with latitude and rating)
correlation = df['Latitude'].corr(df['Aggregate rating'])
print(f"Correlation between Latitude and Rating: {correlation}")
```
