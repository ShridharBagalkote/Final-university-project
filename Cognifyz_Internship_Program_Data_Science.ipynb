```python
# Level-1
 # Task-:  Data Exploration and Preprocessing
 # Task 1
 # Explore the dataset and identify the number of rows and columns.
 # Check for missing values in each column and
 # handle them accordingly.
 # Perform data type conversion if necessary.
 # Analyze the distribution of the target variable
 # ("Aggregate rating") and identify any class
 # imbalances.

import pandas as pd

# 1. Load the Dataset
df = pd.read_csv(r"/Dataset .csv")
df
```

```python
# 2. Explore Rows and Columns
num_rows, num_cols = df.shape
print("Number of rows:", num_rows)
print("Number of columns:", num_cols)
```

```python
# 3. Check for Missing Values
missing_values = df.isnull().sum()
print("\nMissing values per column:\n", missing_values)
```

```python
# 5. Data Type Conversion
#    Inspect data types:
print("\nData Types:\n", df.dtypes)
#    Convert 'Average Cost for two' to float (if it was loaded as a string):
df['Average Cost for two'] = df['Average Cost for two'].astype(float)
```

```python
# 6. Target Variable Distribution
rating_counts = df['Aggregate rating'].value_counts().sort_index()
print("\n'Aggregate rating' distribution:\n", rating_counts)
```

```python
# 7. Identify Class Imbalances
#    Calculate the percentage of each rating class:
class_percentages = rating_counts / len(df) * 100
print("\nClass Percentages:\n", class_percentages)
```

```python
# 8   Visualize the distribution (helpful for spotting imbalances):
import matplotlib.pyplot as plt
plt.bar(class_percentages.index, class_percentages.values)
plt.xlabel('Aggregate Rating')
plt.ylabel('Percentage')
plt.title('Distribution of Aggregate Rating')
plt.show()
```

```python
 #Task:  Geospatial Analysis
 # Task 3
 # Visualize the locations of restaurants on a
 # map using latitude and longitude information.
 # Analyze the distribution of restaurants across different cities or countries.
 # Determine if there is any correlation between the restaurant's location and its rating

import pandas as pd
import plotly.express as px

# Load the dataset
df = pd.read_csv(r"/Dataset .csv")

# Create a scatter mapbox using plotly.express
fig = px.scatter_mapbox(
    df,
    lat="Latitude",
    lon="Longitude",
    hover_name="Restaurant Name",
    hover_data=["Cuisines", "Average Cost for two", "Aggregate rating", "Rating text"],
    color="Rating color",
    color_discrete_map={
        "Dark Green": "green",
        "Green": "lightgreen",
        "Orange": "orange",
        "Red": "red",
        "White": "gray"
    },
    zoom=2,
    height=600,
    width=800
)

# Customize the map appearance
fig.update_layout(mapbox_style="open-street-map")
fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})

# Show the map
fig.show()

# Correlation analysis (example with latitude and rating)
correlation = df['Latitude'].corr(df['Aggregate rating'])
print(f"Correlation between Latitude and Rating: {correlation}")
```

```python
 # Task: Data Visualization
 # Task 3
 # Create visualizations to represent the distribution of ratings using different charts (histogram, barplot, etc.).
 # Compare the average ratings of different cuisines or cities using appropriate visualizations.
 # Visualize the relationship between various
 # features and the target variable to gain insights

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv(r"/Dataset .csv")

# 1. Distribution of Ratings
# Histogram
plt.figure(figsize=(8, 6))
sns.histplot(df['Aggregate rating'], bins=10, kde=True)
plt.title('Distribution of Aggregate Ratings')
plt.xlabel('Aggregate Rating')
plt.ylabel('Frequency')
plt.show()
```

```python
# Box Plot (to show outliers)
plt.figure(figsize=(8, 6))
sns.boxplot(x=df['Aggregate rating'])
plt.title('Distribution of Aggregate Ratings (Box Plot)')
plt.xlabel('Aggregate Rating')
plt.show()
```

```python
# 2. Compare Average Ratings of Cuisines or Cities
# Bar Plot (Average Rating by Cuisine - Top 10)
top_cuisines = df['Cuisines'].value_counts().nlargest(10).index
plt.figure(figsize=(12, 6))
sns.barplot(x='Cuisines', y='Aggregate rating', data=df[df['Cuisines'].isin(top_cuisines)], ci=None)
plt.title('Average Rating of Top 10 Cuisines')
plt.xticks(rotation=45, ha='right')
plt.show()

# Box Plot (to compare distributions with outliers)
plt.figure(figsize=(12, 6))
sns.boxplot(x='Cuisines', y='Aggregate rating', data=df[df['Cuisines'].isin(top_cuisines)])
plt.title('Distribution of Ratings for Top 10 Cuisines (Box Plot)')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# 3. Relationship between Features and Target Variable
# Scatter Plot (Average Cost for two vs. Aggregate rating)
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Average Cost for two', y='Aggregate rating', data=df)
plt.title('Relationship between Average Cost and Rating')
plt.show()
```
